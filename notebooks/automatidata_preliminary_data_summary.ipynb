# Automatidata - Preliminary Data Summary

This work prepares a dataset for future exploratory data analysis (EDA). The purpose is to investigate and understand the data supplied by the NYC Taxi and Limousine Commission. The goal is to leverage Python dataframes in order to perform cursory data inspection and inform team members of the findings. At the end of this stage, the data must be ready to answer questions, yield insights, produce visualizations and be tested through future hypothesis testing and statistical methods.

```python
import pandas as pd
import numpy as np
```

```python
# reading the data
df = pd.read_csv("2017_Yellow_Taxi_Trip_Data.csv")
df.head(10)
```

```python
df.info()
```

The dataset has 22699 samples and does not contain null values. There are a few non-numeric variables, but most of them are numeric. Two of the non-numeric variables contain datetime information.

```python
df.describe()
```

We observe that the maximum value of fare amount is much larger (~$1000) than the 25% - 75% range of values. Additionally, there exist negative values for fare amount in our dataset which is suspicious. Regarding trip distance, while most rides are between 1 - 3 miles, the maximum trip distance is ~34 miles.

```python
# sorting the data by trip distance in descending order
df_sort = df.sort_values(by=["trip_distance"], ascending=False)
df_sort.head(10)
```

```python
# sorting the data by total amount in descending order
desired_columns = ["trip_distance", "fare_amount", "total_amount"]
total_amount_sorted = df.sort_values(["total_amount"], ascending=False)[desired_columns]
total_amount_sorted.head(10)
```

```python
total_amount_sorted.tail(10)
```

For the trip distance, the values align with our earlier data discovery, where we noticed that the longest rides are approximately 34 miles. For the fare amount and the total amount, we observe that the first two values are significantly higher than the others. It should also be noted that the most expensive rides are not necessarily the longest ones.

```python
# determining how many of each payment type are represented in the data
df["payment_type"].value_counts()
```

```python
# calculating the average, maximum and minimum tip for trips paid for with credit card
avg_cc_tip = df[df["payment_type"]==1]["tip_amount"].mean()
max_cc_tip = df[df["payment_type"]==1]["tip_amount"].max()
min_cc_tip = df[df["payment_type"]==1]["tip_amount"].min()
print("Avg credit card tip:", avg_cc_tip)
print("Max credit card tip:", max_cc_tip)
print("Min credit card tip:", min_cc_tip)

# calculating the average, maximum and minimum tip for trips paid for with cash
avg_cash_tip = df[df["payment_type"]==2]["tip_amount"].mean()
max_cash_tip = df[df["payment_type"]==2]["tip_amount"].max()
min_cash_tip = df[df["payment_type"]==2]["tip_amount"].min()
print("Avg cash tip:", avg_cash_tip)
print("Max cash tip:", max_cash_tip)
print("Min cash tip:", min_cash_tip)
```

```python
# determining how many times is each vendor ID represented in the data
df["VendorID"].value_counts()
```

```python
# calculating the mean total amount for each vendor
df.groupby(["VendorID"]).mean(numeric_only=True)[["total_amount"]]
```

```python
# filtering the data for credit card payments only
credit_card = df[df["payment_type"]==1]

# filtering the credit-card-only data for passenger count only
credit_card["passenger_count"].value_counts()
```

```python
# calculating the average tip amount for each passenger count (credit card payments only)
credit_card.groupby(["passenger_count"]).mean(numeric_only=True)[["tip_amount"]]
```

The dataset investigation leads to two variables that are most likely to help build a predictive model for taxi ride fares. These variables are the total amount and the trip distance, as they appear to depict an informative picture of a taxi cab ride.
